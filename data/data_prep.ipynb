### 기상정보

```python
file_list = glob.glob('raw/forecast_1525/forecast_*.csv')
df_list = [pd.read_csv(data_dir + file_path, encoding='euc-kr') for file_path in file_list]
df = pd.concat(df_list, ignore_index=True)
df
```

```python
# 기존 데이터프레임의 컬럼명
column_mapping = {
    '지점': 'station',
    '일시': 'datetime',
    '풍속(m/s)': 'wind_speed',
    '풍향(deg)': 'wind_direction',
    'GUST풍속(m/s)': 'gust_wind_speed',
    '현지기압(hPa)': 'local_pressure',
    '습도(%)': 'humidity',
    '기온(°C)': 'temperature',
    '수온(°C)': 'water_temperature',
    '최대파고(m)': 'max_wave_height',
    '유의파고(m)': 'significant_wave_height',
    '평균파고(m)': 'average_wave_height',
    '파주기(sec)': 'wave_period',
    '파향(deg)': 'wave_direction'
}

# 컬럼명 변경
df_tmp = df.rename(columns=column_mapping)
df_tmp['date'] = pd.to_datetime(df_tmp['datetime']).dt.date
# 결과 확인
print(df_tmp.head())
```

```python
# 집계 함수 정의
agg_funcs = {
    # 'station': None,                     # 고유 식별자, 집계 제외
    # 'datetime': None,                    # 날짜/시간, 집계 제외
    'wind_speed': 'mean',                # 평균 풍속
    # 'wind_direction': 'mean',            # 풍향 제외
    'gust_wind_speed': 'mean',           # 평균 돌풍 풍속
    'local_pressure': 'mean',            # 평균 현지 기압
    'humidity': 'mean',                  # 평균 습도
    'temperature': 'mean',               # 평균 기온
    'water_temperature': 'mean',         # 평균 수온
    'max_wave_height': 'max',            # 최대 파고
    'significant_wave_height': 'mean',   # 평균 유의 파고
    'average_wave_height': 'mean',       # 평균 파고
    'wave_period': 'mean',               # 평균 파주기
    # 'wave_direction': 'mean',            # 파향 제외
}

# 함수 실행
df_agg = create_aggregate_table(df_tmp, index_cols=['station', 'date'], agg_funcs=agg_funcs)
print(len(df_agg["station"].unique()))
df_agg
```

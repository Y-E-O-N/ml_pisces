```python
import pandas as pd
import joblib

model_path = '../../model/test/'
data_path = '../../data/'

# 저장된 모델 및 피처 이름 로드
loaded_model = joblib.load(model_path + 'xgb_광어.joblib')
feature_names = joblib.load(model_path + 'feature_names.pkl')
```

```python
# 데이터 로드
data = pd.read_csv(data_path + 'features/final_oneHot/광어_price_features_notnull.csv')   # 예측에 사용할 데이터셋
data
```

```python
# 데이터 정렬
data['date'] = pd.to_datetime(data['date'])
data = data.sort_values('date')
data
```

```python

# 예측 시작일과 예측 범위 설정
start_date = pd.Timestamp('2025-01-01')  # 첫 번째 예측 날짜
num_days = 1  # 예측할 날짜 수 (이틀 후까지)

# 반복 예측을 위한 데이터프레임 복사
rolling_data = data.copy()

# 예측 결과 저장
predictions = []

for i in range(num_days):
    # 현재 예측 날짜와 이전 날짜 설정
    target_date = start_date + pd.Timedelta(days=i)
    previous_date = target_date - pd.Timedelta(days=1)

    # 이전 날짜 데이터로 피처 구성
    input_data = rolling_data[rolling_data['date'] == previous_date]
    if input_data.empty:
        print(f"데이터가 부족하여 {target_date}를 예측할 수 없습니다.")
        break

    # 필요한 피처만 선택
    X = input_data[feature_names]

    # 예측 실행
    predicted_price = loaded_model.predict(X)[0]  # 예측 값

    # 예측 값을 저장
    predictions.append({'Date': target_date.strftime('%Y-%m-%d'), 'Predicted_Price': predicted_price})

    # 예측 값을 다음날의 lag 피처로 추가
    new_row = {'date': target_date, 'avgPrice': predicted_price}
    rolling_data = pd.concat([rolling_data, pd.DataFrame([new_row])], ignore_index=True)

# 결과 출력
predictions_df = pd.DataFrame(predictions)
print("예측 결과:")
print(predictions_df)

# # CSV로 저장
# predictions_df.to_csv('rolling_predictions.csv', index=False)
# print("예측 결과가 'rolling_predictions.csv'에 저장되었습니다.")
```
